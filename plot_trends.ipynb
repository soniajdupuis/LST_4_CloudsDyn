```python
import xarray as xr
import numpy as np
import rioxarray
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import pymannkendall as mk
from dask.diagnostics import ProgressBar
import numpy as np
import matplotlib.path as mpath
import matplotlib.transforms as mtransform
import cartopy.crs as ccrs
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import matplotlib.transforms as mtransforms
import cartopy.crs as ccrs
```

```python
# --- load datasets ---
files = {
    'DJF': xr.open_dataset('winter_trend.nc'),
    'MAM': xr.open_dataset('spring_trend_night.nc'),
    'JJA': xr.open_dataset('summer_trend_night.nc'),
    'SON': xr.open_dataset('fall_trend_night.nc'),
}



# pick variable name (adjust if needed)
varname = list(files['DJF'].data_vars)[0]

# --- compute common color scale ---
vmin = -2
vmax = 2

vmax = max(abs(vmin), abs(vmax))
vmin = -vmax

proj = ccrs.NorthPolarStereo()

fig, axs = plt.subplot_mosaic(
    [['(a)', '(b)'], ['(c)', '(d)']],
    figsize=(12, 10),
    subplot_kw=dict(projection=proj)
)

labels = ['(a)', '(b)', '(c)', '(d)']
seasons = list(files.keys())

# --- circular boundary ---
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)

mappable = None

# --- plotting loop ---
for season, lab in zip(seasons, labels):
    ax = axs[lab]

    da = files[season][varname]

    mappable = da.plot.pcolormesh(
        ax=ax,
        transform=ccrs.PlateCarree(),
        cmap='coolwarm',
        vmin=vmin,
        vmax=vmax,
        add_colorbar=False
    )

    ax.set_extent([-180, 180, 60, 90], crs=ccrs.PlateCarree())
    ax.set_boundary(circle, transform=ax.transAxes)
    ax.coastlines(linewidth=0.3)

    ax.set_title(season, fontsize=14)

# --- panel labels ---
for lab, ax in axs.items():
    trans = mtransforms.ScaledTranslation(-10/72, 2/72, fig.dpi_scale_trans)
    ax.text(
        0.05, 0.95, lab,
        transform=ax.transAxes + trans,
        fontsize=12,
        verticalalignment='top',
        bbox=dict(facecolor='0.9', edgecolor='none')
    )

#fig.subplots_adjust(right=0.875)

# --- colorbar ---
cbar_ax = fig.add_axes([0.2, -0.1, 0.6, 0.02])

cbar = fig.colorbar(
    mappable,
    cax=cbar_ax,
    orientation='horizontal'
)

cbar.set_label('LST Trend [°C/year]')

plt.tight_layout()
plt.savefig('seasonal_trends_night.png', dpi=300, pad_inches=0.2)
```
